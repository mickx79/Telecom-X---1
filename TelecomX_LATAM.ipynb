#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mickx79/Telecom-X---1/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
```

```python
df.head()
```

```python

# Usamos json_normalize para aplanar las columnas anidadas

df_normalizado = pd.concat([
    df.drop(columns=['customer', 'phone', 'internet', 'account']),  # dejamos las columnas no anidadas
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account'])
], axis=1)

# Ahora df_normalizado tiene todos los datos en un solo nivel
```

```python
df = df_normalizado
df.head()
```

#ðŸ”§ TransformaciÃ³n

```python
# Eliminando las columnas de Churn

df = df[df['Churn'].str.strip() != '']
```

```python
# Cambiando Charges.Total a float


df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
```

```python
# Revisando los vacÃ­os o en blanco
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
# Creando columna 'Cuentas_Diarias'

df['Cuentas_Diarias'] = df['Charges.Monthly'] / 30
print(df[['Cuentas_Diarias','Charges.Monthly']])
```
