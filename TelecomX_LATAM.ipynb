#📌 Extracción

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mickx79/Telecom-X---1/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
```

```python
df.head()
```

```python

# Usamos json_normalize para aplanar las columnas anidadas

df_normalizado = pd.concat([
    df.drop(columns=['customer', 'phone', 'internet', 'account']),  # dejamos las columnas no anidadas
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account'])
], axis=1)

# Ahora df_normalizado tiene todos los datos en un solo nivel
```

```python
df = df_normalizado
df.head()
```

#🔧 Transformación

```python
# Eliminando las columnas de Churn

df = df[df['Churn'].str.strip() != '']
```

```python
# Cambiando Charges.Total a float


df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
```

```python
# Revisando los vacíos o en blanco
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
# Creando columna 'Cuentas_Diarias'

df['Cuentas_Diarias'] = df['Charges.Monthly'] / 30
print(df[['Cuentas_Diarias','Charges.Monthly']])
```

# Informe Final – Proyecto **Telecom X: Análisis de Evasión de Clientes (Churn)**

## 🔹 Introducción

El presente informe corresponde al proyecto **“Telecom X
(parte 1) - Análisis de evasión de clientes”** , como parte del programa **Oracle Next Education (ONE) – Alura Latam**.

La empresa enfrenta una **alta tasa de cancelaciones de clientes**, lo que representa un desafío crítico para su sostenibilidad. El objetivo del proyecto fue **analizar los factores que influyen en la evasión (churn)** a partir de un dataset provisto, con el fin de generar **insights estratégicos** que sirvan como base para futuros modelos predictivos y planes de retención.

---

## 🔹 Limpieza y Tratamiento de Datos

El dataset se obtuvo desde un **repositorio de GitHub** proporcionado por Alura Latam, en formato **JSON**. Para garantizar la correcta manipulación de los datos, se siguieron los siguientes pasos:

1. **Extracción de datos:**

   * Los datos fueron cargados a través de la URL raw de GitHub.
   * Se utilizó el método `json_normalize` de Pandas para **aplanar la estructura** del archivo JSON y obtener un DataFrame utilizable.

2. **Verificación de calidad de datos:**

   * Se comprobó la existencia de valores nulos o duplicados. No se encontraron registros de este tipo.
   * Se identificaron valores de tipo `str` vacíos (`" "`) que entorpecían la lectura, los cuales fueron eliminados.

3. **Transformaciones aplicadas:**

   * Conversión de la columna `Charges.Total` de tipo `str` a **float**, ya que contenía valores numéricos mal tipificados.
   * Creación de una nueva variable: **`Cuentas_Diarias`**, derivada de los registros en la columna `Charges.Monthly`, para facilitar el análisis de comportamiento de gastos diarios.

4. **Variables categóricas:**

   * No se realizó normalización de variables categóricas; se mantuvieron en su formato original para el análisis exploratorio.

Con este tratamiento, se obtuvo un dataset **limpio, consistente y listo** para el análisis exploratorio.

---

## 🔹 Análisis Exploratorio de Datos (EDA)

Se analizaron tanto variables **numéricas** como **categóricas**, utilizando estadísticas descriptivas y visualizaciones. Los principales hallazgos fueron:

1. **Tiempo de permanencia (Tenure):**

   * El **churn es más alto durante los primeros meses** de contrato.
   * A medida que aumenta el tiempo de permanencia, la tasa de abandono disminuye.

2. **Gasto mensual (`Charges.Monthly`):**

   * Existe una tendencia clara: **los clientes con menor gasto mensual presentan mayor probabilidad de abandonar**.
   * Esto se relaciona con usuarios que contratan **planes básicos** y tienen menor compromiso con la compañía.

3. **Método de pago:**

   * Se detectó una fuerte asociación entre el **método de pago Electronic Check** y el abandono.
   * Los clientes que usan tarjetas de crédito o transferencias automáticas muestran una **mayor permanencia**.

4. **Tipo de contrato:**

   * Los contratos **mensuales** concentran la mayoría de los abandonos.
   * Los contratos de **1 año o 2 años** presentan churn mucho más bajo, reflejando un mayor nivel de compromiso con la empresa.

---

## 🔹 Conclusiones e Insights

El análisis permitió obtener los siguientes aprendizajes clave:

* **El churn es más alto al inicio de la relación comercial**, lo que indica que los primeros meses son críticos para la retención.
* **Los clientes de bajo gasto mensual** y con **contratos mensuales** son los más vulnerables a abandonar.
* El **método de pago Electronic Check** es un predictor fuerte de cancelaciones.
* La permanencia se incrementa significativamente con contratos de largo plazo y con métodos de pago automáticos.

---

## 🔹 Recomendaciones Estratégicas

A partir de los hallazgos, se sugieren las siguientes acciones para reducir la evasión:

1. **Mejorar la experiencia de los primeros meses:**

   * Ofrecer beneficios de bienvenida y soporte personalizado.
   * Implementar encuestas tempranas de satisfacción.

2. **Fomentar contratos de mayor permanencia:**

   * Promocionar contratos anuales/bianuales con beneficios exclusivos.
   * Desarrollar programas de fidelización que premien la continuidad.

3. **Optimizar la facturación y métodos de pago:**

   * Incentivar el uso de **pagos automáticos** con tarjeta o débito.
   * Migrar clientes de **Electronic Check** hacia métodos más confiables.

4. **Atender al segmento de bajo gasto mensual:**

   * Diseñar planes personalizados que aumenten el valor percibido.
   * Ofrecer upgrades o servicios complementarios que incentiven la permanencia.
